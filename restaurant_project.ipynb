# Q.1) How many restaurants are there in each city

```python
# plot to see the number of restaurant in city
plt.figure(figsize=(20,10))
sns.countplot(x= restaurant['City'], data=restaurant,palette='icefire')
plt.title('Number of Restaurants in every City',fontsize=20)
plt.xlabel('City',fontsize=15)
plt.ylabel('Count of Restaurants',fontsize=15)
sns.set(style='darkgrid')
plt.show()
```

# Q.3) Which cities have the highest and lowest average rating on indian restaurant 

```python
# average rating of restaurant inevery city
restaurant.groupby('City')['Rating'].mean()
```

```python
# Highest rating in city
restaurant.groupby('City')['Rating'].mean().idxmax()
```

```python
# Lowest rating in city
restaurant.groupby('City')['Rating'].mean().idxmin()
```

- Gurgaon city have the Highest Average rating of the restaurant and kochi city have lowest.

# Q.4) Draw the appropiate plot to see the distribution of Rating and explain your observations.

```python
# Distribution of Rating
sns.histplot(data=restaurant['Rating'],kde=True)
plt.axvline(x=restaurant.Rating.mean(),color='red',alpha=0.5,label='Mean')
plt.axvline(x=restaurant.Rating.median(),color='blue',ls='--',alpha=0.5,label='Median')
plt.legend()
```

- See Maximum number of restaurants rating lies between 4 to 4.5 and average rating is 4.1 and median is 4.2

# create a plot and see the distribution of Rating of every cities explain your observations.

```python
# plot to see the distribution of rating in every cities
rest_rating = restaurant.groupby('City')['Rating'].mean().reset_index()
rest_rating = rest_rating.sort_values('Rating')
plt.figure(figsize=(16,10))
plt.pie(rest_rating['Rating'], labels=rest_rating['City'], autopct='%1.2f%%')
plt.title('Comparison of Average Ratings Across Cities')
plt.show()
```

- see restaurant distribution in every city has almost same 4.06% to 4.62%  maximum restaurans are in Gurgaon and Kochi city have less number.

# Which cities have the most and least expensive restaurant 

```python
# city of most expensive restaurant
most_exp=restaurant.groupby(['City'])['Cost'].max().reset_index()
most_exp.sort_values(by='Cost',ascending=False).head(1)
```

```python
# city of least expensive restaurant
least_exp=restaurant.groupby(['City'])['Cost'].min().reset_index()
least_exp.sort_values(by='Cost').head(1)
```

- Most Expensive restaurant in the city is Mumbai and Least expensive restaurant is in the city of Kolkata.
